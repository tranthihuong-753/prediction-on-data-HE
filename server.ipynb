```python
import numpy as np
from Pyfhel import Pyfhel
```

```python
path_client="/content/drive/MyDrive/Colab Notebooks/Pipeline training & Predict trên dữ liệu mã hóa HE/data/client"
path_server="/content/drive/MyDrive/Colab Notebooks/Pipeline training & Predict trên dữ liệu mã hóa HE/data/server"
```

```python
HE_server = Pyfhel()
HE_server.load_context(f"{path_client}/context.txt")
print(f"1 - {HE_server}")
HE_server.load_public_key(f"{path_client}/public_key.txt")
print(f"2 - {HE_server}")
HE_server.load_relin_key(f"{path_client}/relin_key.txt")
print(f"3 - {HE_server}")
HE_server.load_rotate_key(f"{path_client}/rotate_key.txt")
print(f"4 - {HE_server}")
```

```python
import json
from Pyfhel import PyCtxt
with open(f"{path_client}/data.json", "r") as f:
    data = json.load(f)

Enc_X = [PyCtxt(pyfhel=HE_server, bytestring=bytes.fromhex(enc_str)) for enc_str in data["Enc_X"]]
Enc_y = PyCtxt(pyfhel=HE_server, bytestring=bytes.fromhex(data["Enc_y"]))

print(type(Enc_X[0]))
print(Enc_X[0])
print(type(Enc_y))
print(Enc_y)
```
